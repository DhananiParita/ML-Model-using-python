```python
data = pd.read_csv('/content/std_marks_data.csv')
data
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
data['hours'] = data['hours'].fillna(data['hours'].mean())
data.isna().sum()
```

```python
X = data.iloc[:,:-1]
X
```

```python
Y = data.iloc[:,-1]
Y
```

```python
model = LinearRegression()
model.fit(X,Y)
```

```python
try:
  hours = float(input('How many hours you study? :'))
  age = int(input('What is your age? :'))
  internet = int(input('Do you use internet? : [0=NO and 1=YES]'))


  new_data = [[hours,age,internet]]
  prediction = model.predict(new_data)

  print('I think you will score : ',prediction)

except ValueError:
  print('INVALID INPUT')
```
